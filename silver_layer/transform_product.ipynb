```python
product_df = spark.read.table('de_mini_project.azure_blob_storage.product')
display(product_df)
```

```python
from pyspark.sql import functions as F
product_df = product_df.drop("_file","_line","_modified","_fivetran_synced")
product_df = product_df.withColumnRenamed("@Marked_Price@","marked_price").withColumnRenamed("Product#ID","product_id").withColumnRenamed("[Category_Type]","category").withColumnRenamed("Base_Cost","base_cost")
display(product_df)
```
