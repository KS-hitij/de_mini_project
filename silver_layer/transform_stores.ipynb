```python
stores_df = spark.read.table('de_mini_project.azure_blob_storage.stores')
display(stores_df)
```

```python
from pyspark.sql import functions as F
stores_df = stores_df.drop("_file","_line","_modified","_fivetran_synced")
stores_df = stores_df.withColumnRenamed("ST_ID!!","store_id").withColumnRenamed("Location_Name_&_Address","location").withColumnRenamed("City/Region","city").withColumnRenamed("Mgr_Contact#","manager_contact")
display(stores_df)
```

```python
stores_df.write.format("delta").mode("overwrite").saveAsTable("de_mini_project.silver.stores")
```
