```python
inventory = spark.read.table("de_mini_project.silver.inventory")
product = spark.read.table("de_mini_project.silver.product")
```

```python
display(inventory)
display(product)
```

```python
joint_inventory = inventory.join(product, inventory.sku_id == product.product_id, "inner")
fact_inventory = joint_inventory.drop("product_id","category","last_audit_date")
dim_inventory = joint_inventory.drop("product_id","stock_on_hand","base_cost","marked_price")
display(dim_inventory)
display(fact_inventory)
```

```python
fact_inventory.write.format("delta").mode("overwrite").saveAsTable("de_mini_project.gold.fact_inventory")
dim_inventory.write.format("delta").mode("overwrite").saveAsTable("de_mini_project.gold.dim_inventory")
```
