```python
fact_inventory = spark.read.table("de_mini_project.gold.fact_inventory")
fact_sales = spark.read.table("de_mini_project.gold.fact_sales")
dim_customer = spark.read.table("de_mini_project.gold.dim_customer")
dim_stores = spark.read.table("de_mini_project.gold.dim_stores")
dim_inventory = spark.read.table("de_mini_project.gold.dim_inventory")
dim_returns = spark.read.table("de_mini_project.gold.dim_returns")
```

```python
sales_datacube = fact_sales.join(dim_customer, "customer_id","left").join(dim_stores, "store_id","left").join(dim_inventory, ["sku_id", "store_id"],"left").join(dim_returns, "return_id", "left").dropDuplicates()
display(sales_datacube)
```

```python
datacube_inventory = fact_inventory.join(dim_inventory, ["sku_id", "store_id"],"left").join(dim_stores, "store_id","left")
display(datacube_inventory)
```
